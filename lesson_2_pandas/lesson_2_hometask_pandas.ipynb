```python
import pandas as pd
import numpy as np
```

```python
def answer_one():
  
    energy = pd.read_excel('energy_indicators.xls', skiprows=17, skipfooter=38)
    
    energy = energy.drop(energy.columns[[0, 1]], axis=1)
    energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    
    energy.replace('...', np.nan, inplace=True)
    energy['Energy Supply'] = energy['Energy Supply'] * 1000000
    
    # очищення назв країн
    energy['Country'] = energy['Country'].str.replace(r" \(.*\)", "", regex=True)
    energy['Country'] = energy['Country'].str.replace(r"\d+", "", regex=True)
    energy['Country'] = energy['Country'].str.strip()
    
    energy['Country'] = energy['Country'].replace({
        "Republic of Korea": "South Korea",
        "United States of America": "United States",
        "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
        "China, Hong Kong Special Administrative Region": "Hong Kong"
    })

    #GDP
    GDP = pd.read_excel('world_bank.xls', skiprows=3)
    
    GDP = GDP.rename(columns={GDP.columns[0]: 'Country'})
    GDP.columns = [str(c) for c in GDP.columns]
    
    GDP['Country'] = GDP['Country'].replace({
        "Korea, Rep.": "South Korea", 
        "Iran, Islamic Rep.": "Iran",
        "Hong Kong SAR, China": "Hong Kong"
    })
    
    years = [str(year) for year in range(2006, 2016)]
    GDP = GDP[['Country'] + years]

    #ScimEn
    ScimEn = pd.read_excel('scimagojr.xlsx')
    if 'Country Name' in ScimEn.columns:
        ScimEn = ScimEn.rename(columns={'Country Name': 'Country'})
    
    ScimEn_top15 = ScimEn.head(15)


    df = pd.merge(ScimEn_top15, energy, how='inner', on='Country')
    final_df = pd.merge(df, GDP, how='inner', on='Country')
    
    final_df = final_df.set_index('Country')
    
    columns_to_keep = [
        'Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations',
        'Citations per document', 'H index',
        'Energy Supply', 'Energy Supply per Capita', '% Renewable'
    ] + years

    return final_df[columns_to_keep]


answer_one()
```

```python
def answer_two():
    Top15 = answer_one()
    years = [str(year) for year in range(2006, 2016)]
    avgGDP = Top15[years].mean(axis=1).sort_values(ascending=False)
    return avgGDP.rename('avgGDP')

answer_two()
```

```python
def answer_three():
    Top15 = answer_one()
    avgGDP = answer_two()
    target_country = avgGDP.index[5] 
    
    change = Top15.loc[target_country, '2015'] - Top15.loc[target_country, '2006']
    return change

answer_three()
```

```python
def answer_four():
    Top15 = answer_one()
    Top15['Ratio'] = Top15['Self-citations'] / Top15['Citations']
    max_ratio = Top15['Ratio'].max()
    country = Top15['Ratio'].idxmax()
    return (country, max_ratio)

answer_four()
```

```python
def answer_five():
    Top15 = answer_one()
    Top15['PopEst'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    result = Top15['PopEst'].sort_values(ascending=False).index[2]
    return result

answer_five()
```

```python
def answer_six():
    Top15 = answer_one()
    Top15['PopEst'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Citable docs per Capita'] = Top15['Citable documents'] / Top15['PopEst']
    
    Top15['Citable docs per Capita'] = Top15['Citable docs per Capita'].astype(float)
    Top15['Energy Supply per Capita'] = Top15['Energy Supply per Capita'].astype(float)
    
    return Top15['Citable docs per Capita'].corr(Top15['Energy Supply per Capita'])

answer_six()
```

```python
def answer_seven():
    Top15 = answer_one()
    ContinentDict  = {'China':'Asia', 'United States':'North America', 'Japan':'Asia', 
                      'United Kingdom':'Europe', 'Russian Federation':'Europe', 
                      'Canada':'North America', 'Germany':'Europe', 'India':'Asia',
                      'France':'Europe', 'South Korea':'Asia', 'Italy':'Europe', 
                      'Spain':'Europe', 'Iran':'Asia', 'Australia':'Australia', 
                      'Brazil':'South America'}
    
    Top15['PopEst'] = (Top15['Energy Supply'] / Top15['Energy Supply per Capita']).astype(float)
    Top15['Continent'] = Top15.index.map(lambda x: ContinentDict[x])
    
    result = Top15.groupby('Continent')['PopEst'].agg(['size', 'sum', 'mean', 'std'])
    return result

answer_seven()
```
